## Generate Document abstracts using NLP

### Scraping, parsing the webpage

```python
# scraping using boilerpipe package
%config IPCompleter.greedy=True
import requests
import bs4
from bs4 import BeautifulSoup
res = requests.get('http://radar.oreilly.com/2013/06/the-crypto-currency-ecosystem.html')
type(res)
soup_doc = BeautifulSoup(res.content, 'html.parser')
web_text = soup_doc.get_text()
web_text
```

```python
# Emboiding regular expression 
import re
words = re.findall('\w+', web_text)
words[:20]
```

```python
# Scraping the webdaa using Pyquery and path specified by the Google Selector Gadget API
from pyquery import PyQuery as pq
import requests

url = "http://radar.oreilly.com/2013/06/the-crypto-currency-ecosystem.html"
content = requests.get(url).content
doc = pq(content)
#print(doc)
corpus = doc("p").text()

print(corpus)
```

```python
# Process of detecting sentences from the given paragraph.
content = corpus.split(".")
type(content)
#print(content[0])
```

```python
#NLP applied on the corpus
# Step 1: EOS(End of Sentence) detection
import nltk
#nltk.download('punkt')
# splitting the corpus to sentences
sentences = nltk.tokenize.sent_tokenize(corpus)
sentences
```

```python
# Step2: Word Tokenization
tokens = [nltk.tokenize.word_tokenize(s) for s in sentences]
type(tokens)
```

```python
# Step 3: Word tags description
tags = [nltk.pos_tag(t) for t in tokens]
tags[0]
```

```python
# Step 4: Named entity recognition using Spacy
import spacy
nlp = spacy.load('en')
doc = nlp(corpus)
for ent in doc.ents:
    print(ent.text, ent.label_)

#from spacy import displacy
#spacy.displacy.render(tokens, style='ent', jupyter=True)
#spacy.displacy.render
```

```python
from spacy import displacy
#nlp = spacy.load('custom_ner_model')
doc = nlp(corpus)
spacy.displacy.render(doc, style='ent', jupyter=True)
#displacy.serve(doc, style='ent')
```
